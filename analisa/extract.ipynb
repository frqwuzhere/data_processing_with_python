```python
import requests
import pandas as pd
from io import BytesIO
```

```python
USERNAME = "salma.fitria@tamanzakat.org"
PASSWORD = "Slm21zahr@"

# URL login & download
login_url = "https://donasi.tamanzakat.org/wp-login.php"
download_url = "https://donasi.tamanzakat.org/wp-admin/admin-post.php?action=download_data_donasi&c_id=all&c_date=today&c_range=&status=success"

# Data untuk login form WordPress
payload = {
    'log': USERNAME,
    'pwd': PASSWORD,
    'wp-submit': 'Log In',
    'redirect_to': download_url,
    'testcookie': '1'
}

with requests.Session() as s:
    login = s.post(login_url, data=payload)

    # Cek apakah login berhasil
    if login.status_code == 200 and "logout" in login.text.lower():
        print("✅ Login berhasil.")
        response = s.get(download_url)

        # Cek jenis file
        content_type = response.headers.get("Content-Type", "")
        if "application/vnd.openxmlformats-officedocument.spreadsheetml.sheet" in content_type:
            excel_data = BytesIO(response.content)
            df = pd.read_excel(excel_data, engine="openpyxl")
            print("✅ Berhasil baca DataFrame.")
        else:
            print("⚠️ File bukan Excel. Kemungkinan gagal login / tidak punya akses.")
            print("Status:", response.status_code)
            print("Content-Type:", content_type)
            print("Cuplikan isi:\n",
                  response.content[:500].decode(errors="ignore"))
    else:
        print("❌ Login gagal. Cek username/password.")
```
